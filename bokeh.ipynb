```python
import pandas as pd
from ipywidgets import interact

from bokeh.io import output_notebook, push_notebook, show
from bokeh.plotting import figure
```

```python
colName = ("CODE","TYPE","MATURITY","STRIKE", "RSV", 
          "PUT_CODE", "PUT_PRICE", "PUT_RSV", "PUT_TPRICE", "PUT_VOLATILITY",
          "CALL_CODE","CALL_PRICE","CALL_RSV","CALL_TPRICE","CALL_VOLATILITY",
          "F225_PRICE", "Base_VOL")
df_all = pd.read_csv("ose20190315tp.csv", names=colName)
df_all.set_index("STRIKE",inplace=True)
```

```python
df_all
```

```python
# お試し
@interact(n=5)
def f(n):
    return n
```

```python
output_notebook()
plot = figure(plot_width=600, plot_height=300)
df = df_all[df_all.CODE == "NK225E    "]
df = df[df.MATURITY == 201904]
put_tprice = plot.line(df.index, df["PUT_TPRICE"])
call_tprice = plot.line(df.index, df["CALL_TPRICE"].rolling(10).mean(), color='red')
handle = show(plot, notebook_handle=True)

@interact(n=(0,30))
def redraw(n=10):
    put_tprice.data_source.data = {'x': df.index, 'y': df['PUT_TPRICE']}
    call_tprice.data_source.data = {'x': df.index, 'y': df['CALL_TPRICE'].rolling(n).mean()}
    push_notebook(handle=handle)
```
